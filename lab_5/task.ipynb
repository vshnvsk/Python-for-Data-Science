```python
import pandas as pd
import matplotlib.pyplot as plt

TitanicDataset = pd.read_csv('TitanicDataset.csv', sep=',', header=0)

grouped_data = TitanicDataset.groupby(['Pclass', 'Embarked']).size().unstack()

grouped_data.plot(kind='bar', stacked=True)

plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Class and Embarked Location')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


def a():
    x = list(range(-6, 7))
    y_values = []

    for i in x:
        y = 2 * (i ** 3) - 4 * (i ** 2) + 1
        y_values.append(int(y))

    return x, y_values


def b():
    x = np.linspace(-4 * np.pi, 4 * np.pi, 500)
    y_values = 5 * np.cos(3 * x) - 7 * np.sin(x)
    
    return x, y_values


def c():
    phi = np.linspace(-np.pi/2, np.pi/2, 500)

    rho = 2 + np.cos(phi / 2)

    return phi, rho


fig, axs = plt.subplots(3, 1, figsize=(8, 12))

x_a, y_a = a()
axs[0].plot(x_a, y_a, marker='o', color='b')
axs[0].set_title(r"Plot of $2x^3 - 4x^2 + 1$")
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].grid(True)

x_b, y_b = b()
axs[1].plot(x_b, y_b, color='g')
axs[1].set_title(r"Plot of $5\cos(3x) - 7\sin(x)$")
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].grid(True)

phi_c, rho_c = c()
axs[2] = plt.subplot(313, projection='polar')
axs[2].plot(phi_c, rho_c)
axs[2].set_title(r"$\rho = 2 + \cos(\frac{\phi}{2})$")

plt.tight_layout()
plt.show()
```
