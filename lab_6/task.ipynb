```python
def f(a):
    return a ** 4 - 2 * (a ** 3) - 9 * a + 12

def dy(a):
    return 4 * (a ** 3) - 6 * (a ** 2) - 9

def dy(a):
    h = 1e-10
    return (f(a + h) - f(a)) / h

a0 = 1
alpha = 0.01
epsilon = 0.0000001
print(a0, '\t', f(a0))
print('-' * 17)

while True: 
    a = a0 - alpha * dy(a0)

    if abs(a - a0) <= epsilon:
        break
    else:
        a0 = a
    
    print(round(a, 3), '\t', round(f(a), 3))

print('=' * 100)
print('Значення a, при якому y набуває мінімального значення: ', round(a, 3))
print('Мінімум f: ', round(f(a), 3))
```

```python
import numpy as np

def f(a, b):
    return a ** 2 - 12 * a * b + 4 * (b ** 2) - 4 * a + 3 * b - 1

def grad_f(a, b):
    df_da = 2 * a - 12 * b - 4
    df_db = -12 * a + 8 * b + 3
    return np.array([df_da, df_db])

w0 = np.array([1, 1])  
alpha = 0.01  
epsilon = 1e-3  
max_iterations = 2000  

print(f"Початкове значення: {w0}, f(w0) = {f(w0[0], w0[1])}")
print('-' * 17)

for _ in range(max_iterations):
    grad = grad_f(w0[0], w0[1])
    w = w0 - alpha * grad

    if np.linalg.norm(grad) < epsilon:
        break

    w0 = w

    print(np.round(w, 5), '\t', np.round(f(w[0], w[1]), 5))

print(f"Мінімум знаходиться в точці: {w}, f(w) = {f(w[0], w[1])}")
```
